```python
from pydub import AudioSegment
audio_file = AudioSegment.from_wav("only_speech_2016a.wav")  # Replace with your audio file path and format
Total_msec = len(audio_file)
print(Total_msec)
# Iterate through speech segments
i=0;
start_time=0
start_time = start_time # Convert to milliseconds
end_time = 20 * 1000  # Convert to milliseconds
while(True):
    print(start_time,end_time)
    # start_time = start_time # Convert to milliseconds
    # end_time = 20 * 1000  # Convert to milliseconds
    

    # Segment audio file
    if(end_time>Total_msec):
      print("end : ",start_time,Total_msec)
      segmented_audio = audio_file[start_time:Total_msec]
      
      output_filename = f"lastpiece_2016a_{i}.wav"  # Generate output filename based on speaker ID
      segmented_audio.export(output_filename, format="wav")  # Save segmented audio as WAV file
      generate_rttm(output_filename)
      print(f"Segmented audio for saved as '{output_filename}'")
      break
    segmented_audio = audio_file[start_time:end_time]
    # Save segmented audio sample
    output_filename = f"segmAudio_2016a_{i}.wav"  # Generate output filename based on speaker ID
    segmented_audio.export(output_filename, format="wav")  # Save segmented audio as WAV file
    print(f"Segmented audio for saved as '{output_filename}'")
    generate_rttm(output_filename)
    i=i+1
    start_time = end_time
    end_time += 20*1000
    # print(start_time,end_time)
print("Segmentation complete!")
```
